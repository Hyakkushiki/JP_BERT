Prepare a pre-trained Japanese BERT model

```python


!pip install transformers==3.0.2

```

```python
import torch
from transformers.modeling_bert import BertModel
from transformers.tokenization_bert_japanese import BertJapaneseTokenizer
```

Define the tokenizer. We will use the cl-Tohoku model and tokenizer.

```python
tokenizer = BertJapaneseTokenizer.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
```

Here you can see the details of our BERT model.

```python
model = BertModel.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
print(model)
```

Here you can see model details, such as the activation function of the hidden laters (gelu). Each vocabulary term is a 768-dimension vector. Up to 512 characters (subwords) can be handled at once.

```python
from transformers import BertConfig

config_japanese = BertConfig.from_pretrained('cl-tohoku/bert-base-japanese-whole-word-masking')
print(config_japanese)
```

Here is a demonstration of how pytorch/BERT assign an id number to vocabulary so that math can be performed on words. Remember that each word is a 768-dimension vector, but BERT cannot handle so many numbers at one time. The character ”を”, for example, has an id of "811", which points to a list in the tokenizer dictionary that contains a 768-dimension vector representing ”を” in the Japanese language.

```python
text1 = "会社をクビになった。"
text2 = "テレワークばかりでクビが痛い。"
text3 = "会社を解雇された。"
```

Note that the sentences start with a [CLS] token and end with a [SEP] token. This is the BERT format and is true for every input string.

```python
input_ids1 = tokenizer.encode(text1, return_tensors='pt')  #pt stands for "pytorch"

print(tokenizer.convert_ids_to_tokens(input_ids1[0].tolist()))  # print the Japanese text
print(input_ids1)  # print the corresponding ids for the characters
```

Repeat for text2

```python
input_ids2 = tokenizer.encode(text2, return_tensors='pt')  

print(tokenizer.convert_ids_to_tokens(input_ids2[0].tolist()))  
print(input_ids2)  
```

Repeat for text3

```python
input_ids3 = tokenizer.encode(text3, return_tensors='pt') 

print(tokenizer.convert_ids_to_tokens(input_ids3[0].tolist()))  
print(input_ids3)
```

Next, input our text (converted to ids) into BERT. Let's try text1. As you can see, the result is a tensor (a matrix of matrices).

```python
result1 = model(input_ids1)

print(result1)
```

We can get more general information about our tensor result with with .shape() function. Also note that result1[0] represents the sequence output, and result1[1] represents the pooled output.

```python
print(result1[0].shape)
print(result1[1].shape)
```

Result1[0] contains 1 row, 9 columns... and 768 elements in a third dimension. "1" represents the number of batches, "9" represents the 9 ids that we fed into BERT from text1, and "768" represents the dimensions of each input id. 
Result1[1] contains 1 row and 768 colums (or rather, a list of 768 numbers). 

```python
print(result1[0])
```

To make sure that our dimensions are correct, and to see the raw output from out BERT model, we can simply print the pooled output, which is result1[1]. Also note that the hyperbolic tangent function (tanh) is the activation function used in the final ouput layer.

```python
print(result1[1])
```

Now, let's feed text2 and text3 into BERT and get result2 and result3.

```python
result2 = model(input_ids2)
result3 = model(input_ids3)
```

Each word in BERT is represented by a 768-dimension vector. Before we fed the text into BERT, those vectors were determined by the dictionary/tokenizer. After we fed the text through BERT, the 768 dimensions of each vector (word) was changed! 
Now, we can compare the outputs for each sentence and do some mathematical comparisons. 

Let's consider two words in our three sentences: "クビ" and "解雇". First, we will identify the target words in our three results.

```python
word_vec1 = result1[0][0][3][:]  # ”クビ” in text1 (id #3)
word_vec2 = result2[0][0][5][:]  # ”クビ” in text2（id #5）
word_vec3 = result3[0][0][3][:]  # ”解雇” in text3（id #3）
```

We can confirm that the vectors for ”クビ” have changed based on the surrounding words in its sentences (in other words, the vectors have changed based on the BERT model and the input sentence context.)

```python
print("Output 'クビ' from Text 1"+'\n')
print(word_vec1)
print("Output 'クビ' from Text 2"+'\n')
print(word_vec2)
```

Now we want to answer the question: 

"Is the meaning of ”クビ” in Text 1 more similar to the meaning of ”クビ” in Text 2? Or, is the meaning of ”クビ” in Text 1 more similar to the meaning of ”解雇” in Text 3?"

We can answer this question many ways. One way is to use Cosine Similarity, which follows the formula: 

similarity=max(∥x1​∥2​⋅∥x2​∥2​,ϵ)x1​⋅x2​​

```python
cos = torch.nn.CosineSimilarity(dim=0)
cos_sim_12 = cos(word_vec1, word_vec2)
cos_sim_13 = cos(word_vec1, word_vec3)

print(cos_sim_12)
print(cos_sim_13)
```

The cosine similarity between ”クビ” in text1 and text2 is 0.66, whereas the similarity between ”クビ” in text1 and ”解雇” in text3 is 0.78. 

Therefore, the meaning of ”クビ” in text1 and the meaning of ”解雇” in text3 is more similar than the meanings of ”クビ” in text1 and text2. 

This is correct, as the ”クビ” in text1 refers to getting fired from a job (”解雇”) not the anatomical "neck" meaning in text2.
